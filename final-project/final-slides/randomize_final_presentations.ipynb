# Student Final Presentation Order Randomizer

This notebook randomizes the order of student presentations for the final project presentations.

## Instructions

1. Set the random seed in the cell below (for reproducibility)
2. Run all cells to generate the randomized presentation order
3. The output will show a table with presentation order, student names, and emails


```python
import pandas as pd
import numpy as np
```

## Set Random Seed

Change the seed value below to get different randomizations. Using the same seed will always produce the same order.

```python
# Set your random seed here
RANDOM_SEED = 2026

# Set the seed for reproducibility
np.random.seed(RANDOM_SEED)

print(f"Random seed set to: {RANDOM_SEED}")
```

## Generate Randomized Presentation Order

```python
# Load the roster CSV file
roster_df = pd.read_csv('../../Spring_2026-STATS_roster.csv')

# Shuffle the dataframe
randomized_df = roster_df.sample(frac=1).reset_index(drop=True)

# Create presentation order table
presentation_order = pd.DataFrame({
    'Presentation #': range(1, len(randomized_df) + 1),
    'Name': randomized_df['Name'],
    'Email': randomized_df['E-mail']
})

# Display the table
presentation_order
```
